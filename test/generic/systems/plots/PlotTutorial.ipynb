This Notebook creates plots of the timings, speedups and efficiencies for the various ScaFaCoS methods against the number of cores for a given testcase with a certain number of charges and a given tolerance. With the benchmark module it is also easily possible to plot against the tolerance or against the number of charges.<p>    First we define some constants that will be used later. If you want to plot other data, just modify these.

```python
testcase='3d-periodic/bm_cloud_wall.xml'charges=8100tolerance=1e-3
```

The following will load the data from the given testcase via the module `benchmarks.py`.

```python
import benchmarksall_charges, all_tolerances, all_cores, timing, speedup, efficiency = benchmarks.read(testcase)
```

The function first returns the different labels of the axes, namely the charges, the tolerances, and the cores.

```python
print all_charges, all_tolerances, all_cores
```

The rest of the returned data are dictionaries. The keys of the dictionaries are the method names:

```python
print timing.keys()
```

The values of the dictionary are 3D-Numpy arrays containing the timing data in the 3 axes. The different axes are:<ol>    <li>number of charges</li>    <li>tolerance</li>    <li>number of cores</li></ol>

```python
print type(timing['p3m'])print timing['p3m'].shape
```

Now we can easily access arbitrary benchmark data, e.g. the timings of p3m for the given number of charges and tolerance with 512 cores:

```python
ixcha = all_charges.index(charges)ixtol = all_tolerances.index(tolerance)ixcor = all_cores.index(512)print timing['p3m'][ixcha,ixtol,ixcor]
```

Like this we can plot different timings against each other:

```python
loglog(all_cores, timing['p3m'][ixcha,ixtol,:], 'o-')loglog(all_cores, timing['fmm'][ixcha,ixtol,:], 'o-')legend(('p3m', 'fmm')); xlabel('#cores'); ylabel('Timing [s]')show()
```
